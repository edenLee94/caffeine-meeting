````python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens')
````

# BERT 이용

````python
df['BERT_BASE'] = df['evaluation'].map(lambda x: model.encode(str(x))) 
````

````python
from numpy import dot
from numpy.linalg import norm

# cos 유사도를 함수로 구현해준다.
def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B)) 
````

````python
df['BERT_BASE']
````

````python
caf_df = df.groupby(['cafe_name', 'region', 'evaluation'])['BERT_BASE'].mean().reset_index()
````

````python
caf_df
````

````python
def lecture_searching(keyword, data = caf_df):
  boolean_index = df['cafe_name'].map(lambda x : keyword in x)
  tmp_df = df[boolean_index]

  if len(tmp_df) == 1: 
    index = tmp_df.index[0]
  else:
    print(tmp_df[['cafe_name', 'region']])
    index = int(input('인덱스를 입력하세요: '))
  
  target = df['BERT_BASE'][index]

  df['score'] = df.apply(lambda x: cos_sim(x['BERT_BASE'] , target), axis = 1)
  return df[['cafe_name','region','evaluation','score']].sort_values(by = 'score', ascending = False)[0:11]
````

pickle은  list, dict와 같은 파이썬 객체를 그 형태 그대로 저장하고, 불러올 수 있게끔 하는 패키지

'객체A'라는 이름의 파이썬 객체를 '파일A.pickle'이라는 피클 확장자의 파일로 저장해 줍니다. as 다음에 등장하는 f는 '파일 핸들'이므로 다른 이름을 써도 상관이 없습니다. 

pickle.dump()를 사용하여 해당 내용을 path(경로) 파일에 입력

#### pickle.HIGHEST_PROTOCOL
정수, 사용 가능한 가장 높은 프로토콜 버전. 이 값은 함수 dump()와 dumps() 그리고 Pickler 생성자에 protocol 값으로 전달될 수 있습니다.
