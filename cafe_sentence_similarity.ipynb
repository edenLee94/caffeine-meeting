#### 리뷰 데이터에 대한 자연어처리를 위해서 데이터 전처리를 실시해준다.

```python
import re
```

re.compile('[^가-힣]')을 사용할 때 string으로 변환해줘야합니다.

```python
def extract_word(text):
    hangul = re.compile('[^가-힣]')
    result = hangul.sub(' ', text) 
    return result
```

```python
len(df)
```

```python
extract_word(df['evaluation'][19])
```

```python
extract_word(df['evaluation'][19])
' '.join(df['evaluation'][19].split())
```

```python
df['evaluation'] = df['evaluation'].apply(lambda x : extract_word(str(x)))
```

```python
from hanspell import spell_checker
for i in range(1,len(df)):
    spelled_sent = spell_checker.check(df['evaluation'][i])
    spell = spelled_sent.checked
    df['evaluation'][i] = spacing(spell)
    print(df['evaluation'][i])
```

```python
for i, document in enumerate(df['evaluation']):
    clean_words = []
    for word in okt.pos(document, stem=True): #어간 추출
        if word[1] in ['Noun', 'Verb','Adjective']: #명사, 동사, 형용사
            clean_words.append(word[0])
    # print(clean_words)
    document = ' '.join(clean_words)
    print(document) 
    df['evaluation'][i] = document
```

```python
vectorize = TfidfVectorizer(min_df = 3, ngram_range = (1,3))
```

```python
features = vectorize.fit_transform(df['evaluation'])
feature_vect_dense = features.todense()
# pd.DataFrame(feature_vect_dense, columns = vectorize.get_feature_names_out())
```

```python
cos_sim = cosine_similarity(features, features, dense_output=True)
```
