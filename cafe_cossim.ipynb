```python
import re
```

re.compile('[^가-힣]')을 사용할 때 string으로 변환해줘야합니다.

```python
def extract_word(text):
    hangul = re.compile('[^가-힣]') 
    result = hangul.sub(' ', text) 
    return result
```

```python
extract_word(df['evaluation'][19])
### 띄어쓰기 고쳐야할 것 같다.
```

```python
df['evaluation'] = df['evaluation'].apply(lambda x : extract_word(str(x)))
```

```python
from konlpy.tag import Okt
from konlpy.tag import Mecab
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
okt = Okt()
```

```python
vectorizer = TfidfVectorizer(min_df = 3, ngram_range = (1,2), tokenizer = okt.phrases)
```

```python
features = vectorizer.fit_transform(df['evaluation'])
pd.DataFrame(features.todense(), columns = vectorizer.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim = cosine_similarity(features, features, dense_output=True)
```

```python
pd.DataFrame(cos_sim)
```

```python
index = pd.DataFrame(cos_sim)[5].sort_values(ascending = False)[:11].index
```
