```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

AI-hub site에서 웰니스 대화 스크립트 데이터셋을 신청 후 xlsx파일을 csv로 변환하여 사용하기

```python
xlsx = pd.read_excel("/content/drive/MyDrive/datagit/웰니스_대화_스크립트_데이터셋.xlsx")
xlsx.to_csv("/content/drive/MyDrive/datagit/wellness_dataset_original.csv")
```

```python
df = pd.read_csv('/content/drive/MyDrive/datagit/wellness_dataset_original.csv')

df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
df.loc[0, '유저']
```

```python
df = df[-df['챗봇'].isna()]
```

https://huggingface.co/jhgan/ko-sroberta-multitask

문장 및 단락을 768차원 밀집 벡터 공간에 매핑하고 클러스터링 또는 의미 검색과 같은 작업에 사용할 수 있는 문장 변환기 모델을 사용한다.

 -> ko-sroberta-multitask

https://huggingface.co/jhgan/ko-sroberta-multitask

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')
model.encode(df.loc[0,'유저'])
```

# 유저 대화내용 인코딩

```python
df['embedding'] = pd.Series([[]] * len(df))

df['embedding'] = df['유저'].map(lambda x: list(model.encode(x)))
df.head()
```

```python
df.to_csv('wellness_dataset.csv', index=False)
```

```python
text = '요즘 머리가 아프고 목도 뻐근하고 힘들어요'

embedding = model.encode(text)
```

```python
df['similary'] = df['embedding'].map(lambda x:cosine_similarity([embedding], [x]).squeeze())

df.head()
```

```python
answer = df.loc[df['similary'].idxmax()]

print('구분', answer['구분'])
print('유사한 질문', answer['유저'])
print('챗봇 답변', answer['챗봇'])
print('유사도', answer['similary'])
```
